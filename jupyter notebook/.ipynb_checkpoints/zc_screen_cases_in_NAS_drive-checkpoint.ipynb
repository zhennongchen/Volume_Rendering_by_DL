```python
import os
import numpy as np
import glob
import nibabel as nib
import matplotlib.pyplot as plt
import zc_function_list as ff
import pandas as pd
```

```python
def find_function_folder(folder_list,keywordlist):
    # this function find the function folder that contain the keyword in the keyword list.
    # default rank of keywordlist is "Earliest_to_Latest" > "CCTA/CTA" > something else
    
    total_round = len(keywordlist)
    t = 0
    while 1==1:
        assess = [keywordlist[t] in F for F in folder_list]
        if True in assess:
            return t,keywordlist[t]
        else:
            t += 1
        if t == len(keywordlist):
            word = 'Else'
            return t,word
```

```python
def read_DicomDataset(dataset,elements):
    result = []
    for i in elements:
        if i in dataset:
            result.append(dataset[i].value)
        else:
            result.append('')
    return result
```

```python
def massive_list_append(list_list,append_list):
    # to minimize the line of codes for list append
    assert len(list_list) == len(append_list)
    [list_list[i].append(append_list[i]) for i in range(0,len(append_list))]
```

```python
years = ff.find_all_target_files(['2019'],main_path)
```

```python
patients = ff.find_all_target_files(['CVC1902*'],years[0])
```

```python
keywordlist = ['_TO_','_to_','CCTA','CTA','ccta','cta','Function','function']
dicom_parameter_list = ['Manufacturer','ManufacturerModelName','StudyDescription','PatientSex','PatientAge','ProtocolName']
```

```python
patient_id = []
scan_year = []
function = []
directories = []
directories_sub = []

Dicom = []
Manufacturer = []
ModelName = []
StudyDescription = []
Sex = []
Age = []
Protocol = []

for patient in patients:
    patient_id.append(os.path.basename(patient))
    scan_year.append(os.path.basename(os.path.dirname(patient)))
    print(os.path.basename(patient))
    D = [os.path.basename(os.path.join(patient, o)) for o in os.listdir(patient) \
         if os.path.isdir(os.path.join(patient,o))]
    
    # we only want the patients that has directories for cardiac function
    # the hypothesis here is that those patients will have directory name with "%"
    function_D = []
    count = 0
    for i in D:
        if '%'in i:
            count += 1
            function_D.append(i)
            
    if count >= 5:
        function.append('Yes')
        directories.append(function_D)
        # select the function directory (in case there are more than one series of function folders)
        t,word = find_function_folder(function_D,keywordlist)
        if t < len(keywordlist):# not 'Else'
            function_D_sub = []
            [function_D_sub.append(ii) for ii in function_D if word in ii]
        else:
            function_D_sub = []
            [function_D_sub.append(ii) for ii in function_D]
        directories_sub.append(function_D_sub)
             
        
    else:
        function.append('NO')
        directories.append(D);
        directories_sub.append(' ');
    
    # get dicom info
    if count >= 5:
        dicom_list = ff.find_all_target_files(['*.dcm'],os.path.join(patient,function_D_sub[0]))
        if len(dicom_list) > 0:
            Dicom.append('Yes')
            read = pydicom.read_file(dicom_list[0])
            data_list = read_DicomDataset(read,dicom_parameter_list)
        else:
            Dicom.append('No')
            data_list = ['']*len(dicom_parameter_list)
    else:
        Dicom.append('')
        data_list = ['']*len(dicom_parameter_list)
    
    massive_list_append([Manufacturer,ModelName,StudyDescription,Sex,Age,Protocol],data_list)
print('finish')
```
