```python
class Num:
    def __init__(self, n1, n2):
        self.firstnum = n1
        self.secondnum = n2
        
    def printname(self):
        print(self.firstnum, self.secondnum)

class Student(Num):
    def __init__(self,n1,n2):
        super().__init__(n1, n2)
        
   
    def printname(self):
        print(self.firstnum+1, self.secondnum+2)
    
    
x = Num(1,2)
x.printname()

s = Student(1,2)
s.printname()
```
